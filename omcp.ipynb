```python
def omega_func(kappa,N,R,alpha,xx1,xn,rn,pars):
    alph_frac = (1-alpha)/(1+alpha)
    omega = np.zeros((N,N))
    # Call derivatives of a_n's
    a_nn_prime = np.zeros((N+1,N+1))
    a_nn_primeprime = np.zeros((N+1,N+1))
    for i in range(0,N+1):
        a_nn_prime[i,:] = a_n.an_func(rn[i],N,R)[1,:]
        a_nn_primeprime[i,:] = a_n.an_func(rn[i],N,R)[2,:]
        
    for m in range(1,N+1):
#         kappa = kappa_func(xx1,0,r1,r2,zlist,rn,pars)[0]
        for n in range(1,N+1):
            omega[m-1,n-1] = (-a_nn_prime[0,n]/(3*kappa*alph_frac/2+a_nn_prime[0,0])*(1/rn[m]*a_nn_prime[m,0]+a_nn_primeprime[m,0])+1/rn[m]*a_nn_prime[m,n]+a_nn_primeprime[m,n])
#             omega[m-1,n-1] = 1/rn[m]*a_nn_prime[m,n]+a_nn_primeprime[m,n]
    return omega 
```
